```python
 # !pip install requests beautifulsoup4 pandas
from bs4 import BeautifulSoup
import requests
import concurrent.futures
import pandas as pd
import re

def clean_song_name(raw_name):
    # Remove smart/regular quotes at start and end, keep [ft. ...]
    cleaned = re.sub(r'^[“"]+', '', raw_name)
    cleaned = re.sub(r'[”"]+\s*(?=\[ft\.)', '', cleaned)
    cleaned = re.sub(r'[”"]+$', '', cleaned)
    return cleaned.strip()


def scrape_website(url):
    try:
        site = requests.get(url, timeout=10)
        site.raise_for_status()
        soup = BeautifulSoup(site.text, 'html.parser')

        bill_cards = soup.find_all('div', class_="body__inner-container")

        rankings, introductions, artists, song_names, pics = [], [], [], [], []

        auto_rank = 100  # Start from 100

        # Extract year from URL
        match = re.search(r'(\d{4})', url)
        year = match.group(1) if match else "Unknown"


        for item in bill_cards:
            for h2 in item.find_all('h2'):
                p = h2.find_next_sibling("p")
                if h2 and p:
                    h2_text = h2.get_text(strip=True)
                    p_text = p.get_text(strip=True)

                    try:
                      artist, song_name = h2_text.split(": ", 1)
                      artist = [a.strip() for a in artist.split(" / ")]  # 🔥 split multi-artist
                      song_name = clean_song_name(song_name)  # ✅ use the smart quote cleaner
                    except ValueError:
                      artist, song_name = None, None

                    img_url = None
                    img_wrapper = h2.find_previous("div", class_="GenericCalloutWrapper-tojWn ksjPnB")
                    if img_wrapper:
                        img = img_wrapper.find("img")
                        if img and img.has_attr("src"):
                          img_url = img["src"]

                    rankings.append(str(auto_rank))
                    auto_rank -= 1
                    introductions.append(p_text)
                    artists.append(artist)
                    song_names.append(song_name)
                    pics.append(img_url)


        return {
            'url': url,
            'year': year,
            'total': len(rankings),
            'rankings': rankings,
            'artists': artists,
            'songs': song_names,
            'introductions': introductions,
            'images': pics
        }

    except Exception as e:
        return {'url': url, 'error': str(e)}

if __name__ == "__main__":
    urls = [
        'https://pitchfork.com/features/lists-and-guides/best-songs-2024/',
        'https://pitchfork.com/features/lists-and-guides/best-songs-2023/',
        'https://pitchfork.com/features/lists-and-guides/best-songs-2022/',
        'https://pitchfork.com/features/lists-and-guides/best-songs-2021/',
        'https://pitchfork.com/features/lists-and-guides/best-songs-2020/',
        'https://pitchfork.com/features/lists-and-guides/best-songs-2019/',
        'https://pitchfork.com/features/lists-and-guides/the-100-best-songs-of-2018/',
        'https://pitchfork.com/features/lists-and-guides/the-100-best-songs-of-2017/',
    ]

    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(scrape_website, urls))

    all_rows = []
    for result in results:
        print(f"\n--- {result['url']} ---")
        if 'error' in result:
            print("❌ Error:", result['error'])
        else:
            print(f"✅ Scraped {result['total']} entries.")
            for i in range(result['total']):
                all_rows.append({
                    'Year': result['year'],
                    'Rank': result['rankings'][i],
                    'Artist': result['artists'][i],
                    'Song': result['songs'][i],
                    'Introduction': result['introductions'][i],
                    'Image': result['images'][i]
                })

    df = pd.DataFrame(all_rows)
    df.to_csv("pitchfork_best_songs_2015_2024_years.csv", index=False)
    print("\n✅ Saved to pitchfork_best_songs_2015_2024_years.csv")
    print(df.head())
```

```python
def scrape_old_pitchfork_format(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        # Extract year from URL
        matches = re.findall(r'\d{4}', url)
        year = matches[-1] if matches else "Unknown"

        results = []

        for h1_div in soup.find_all("div", class_="heading-h1"):
            img = h1_div.find_previous("img")
            if img:
                src = img.get("src") or img.get("data-src")

        for h3_div in soup.find_all("div", class_="heading-h3"):
            try:
                # Get ranking and introduction from p tags following heading-h3
                p_rank = h3_div.find_next_sibling("p")
                p_intro = p_rank.find_next_sibling("p") if p_rank else None

                # Get song (h2) and artist (div.heading-h1) by walking back
                song_tag = h3_div.find_previous("h2")
                artist_tag = h3_div.find_previous("div", class_="heading-h1")

                if not (artist_tag and song_tag and p_rank and p_intro):
                    continue

                artist = artist_tag.get_text(strip=True)
                song = song_tag.get_text(strip=True).strip('“”"\'')
                rank = p_rank.get_text(strip=True)
                intro = p_intro.get_text(strip=True)
                url = h3_div.find_previous("a")["href"] if h3_div.find_previous("a") else None

                results.append({
                    'Year': year,
                    'Rank': rank,
                    'Artist': artist,
                    'Song': song,
                    'Introduction': intro,
                    'Source URL': url,
                    'Image': src
                })

            except Exception as e:
                print(f"⚠️ Skipped entry {i} due to error: {e}")
                continue

        return results

    except Exception as e:
        print(f"❌ Failed to scrape {url}: {e}")
        return []

# 🎯 Target URLs for 2015–2016 format
urls = [
    'https://pitchfork.com/features/lists-and-guides/9981-the-100-best-songs-of-2016/',
    'https://pitchfork.com/features/lists-and-guides/9765-the-100-best-tracks-of-2015/'
]

# Run scraping and gather results
all_entries = []
for url in urls:
    all_entries.extend(scrape_old_pitchfork_format(url))

# Load existing file (create empty if not found)
try:
    df_existing = pd.read_csv("pitchfork_best_songs_2015_2024_years.csv")
except FileNotFoundError:
    df_existing = pd.DataFrame()

# Create DataFrame from new entries
df_new = pd.DataFrame(all_entries)

# Combine both
df_combined = pd.concat([df_existing, df_new], ignore_index=True)

# Save back to the same file
df_combined.to_csv("pitchfork_best_songs_2015_2024_years.csv", index=False)

print("\n✅ Appended to pitchfork_best_songs_2015_2024_years.csv")
print(df_combined.tail())
```

```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd


# Scrape a single Billboard year-end list
def scrape_billboard_year(url):
    year_match = re.search(r'(\d{4})', url)
    year = year_match.group(1) if year_match else "Unknown"

    resp = requests.get(url)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")

    table = soup.find("table", {"class": "wikitable"})
    records = []

    for tr in table.find_all("tr")[1:]:
        cols = tr.find_all(["th", "td"])
        if len(cols) < 3:
            continue

        rank = cols[0].get_text(strip=True)

        # Title + Link
        raw_title = cols[1].get_text(strip=True).strip('"')
        raw_title_link = cols[1].find("a")
        title_link = None
        if raw_title_link:
            raw_title = raw_title_link.get_text(strip=True)
            if raw_title_link.has_attr("href"):
                title_link = "https://en.wikipedia.org" + raw_title_link["href"]

        # Artist
        raw_artist = cols[2].get_text(strip=True)

        records.append({
            'year': year,
            'billboard_rank': rank,
            'title': clean_title(raw_title),
            'title_link': title_link,
            'artists': split_artists(raw_artist),
        })

    return records

# URLs to scrape
urls = [
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2015',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2016',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2017',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2018',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2019',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2020',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2021',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2022',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2023',
    'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2024',
]

# Aggregate results across all years
all_records = []
for url in urls:
    all_records.extend(scrape_billboard_year(url))

# Convert to DataFrame and export
df = pd.DataFrame(all_records, columns=["year", "billboard_rank", "title",'title_link', "artists"])
print(f"{len(df)} rows scraped.")
df.to_csv("billboard_top_songs_2015_2024.csv", index=False)
print("billboard.csv written!")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

# Load merged dataset
df = pd.read_csv("merged_2015_2024.csv")

# Define the image scraper
def get_wikipedia_cover_image(wiki_url):
    if pd.isna(wiki_url):
        return None
    try:
        time.sleep(1)
        res = requests.get(wiki_url, timeout=10)
        soup = BeautifulSoup(res.text, 'html.parser')
        infobox = soup.find('table', class_='infobox')
        img_tag = infobox.find('img') if infobox else None
        return 'https:' + img_tag['src'] if img_tag and img_tag.get('src') else None
    except Exception as e:
        print(f"❌ Error fetching {wiki_url}: {e}")
        return None

# Apply the scraper to rows where image is missing
df["image"] = df.apply(
    lambda row: get_wikipedia_cover_image(row["title_link"]) if pd.isna(row["image"]) else row["image"],
    axis=1
)

# Save updated dataset
df.to_csv("merged_2015_2024_with_wiki_images.csv", index=False)
print("✅ Cover image links updated and saved.")
```

```python
import pandas as pd
import requests

# Load your merged CSV
df = pd.read_csv("merged_2015_2024.csv")

# Add new column for lyrics
df["lyrics"] = None

# Loop through each row and fetch lyrics
for i, row in df.iterrows():
    song = row["song"]
    artists = row["artists"]

    # Use only the first artist
    if pd.isna(song) or pd.isna(artists):
        continue
    first_artist = str(artists).split(",")[0].split("/")[0].strip()

    # Prepare API URL
    api_url = f"https://api.lyrics.ovh/v1/{first_artist}/{song}"

    try:
        response = requests.get(api_url, timeout=10)
        if response.status_code == 200:
            df.at[i, "lyrics"] = response.json().get("lyrics")
        else:
            print(f"❌ No lyrics found for: {first_artist} - {song}")
    except Exception as e:
        print(f"⚠️ Error fetching: {first_artist} - {song} → {e}")

# Save updated file
df.to_csv("merged_2015_2024_with_lyrics.csv", index=False)
print("✅ Lyrics added and saved to merged_2015_2024_with_lyrics.csv")
```

```python
# top_artists_interactive_bar.py

import pandas as pd
import re
from ast import literal_eval
import plotly.graph_objects as go

# Load dataset
df = pd.read_csv("/content/merged_2015_2024_with_wiki_images.csv")
df = df.dropna(subset=["artists"])

artist_song_map = {}

for _, row in df.iterrows():
    raw = row["artists"]
    song = row["song"]
    try:
        parsed = literal_eval(raw) if raw.startswith("[") else [raw]
    except:
        parsed = [raw]

    for entry in parsed:
        names = re.split(r"\s*,\s*|\s*/\s*|\s+and\s+", entry)
        for name in names:
            artist = name.strip()
            if artist:
                if artist not in artist_song_map:
                    artist_song_map[artist] = {"count": 0, "songs": set()}
                artist_song_map[artist]["count"] += 1
                artist_song_map[artist]["songs"].add(song)

# Top 10 artists
sorted_artists = sorted(artist_song_map.items(), key=lambda x: x[1]["count"], reverse=True)[:10]

artists = [a[0] for a in sorted_artists]
counts = [a[1]["count"] for a in sorted_artists]
hover_texts = [f"{a[0]}<br>Count: {a[1]['count']}<br>Songs: {', '.join(a[1]['songs'])}" for a in sorted_artists]

fig = go.Figure(go.Bar(
    x=artists,
    y=counts,
    hovertext=hover_texts,
    hoverinfo="text",
    marker=dict(color="#a56eff")
))

fig.update_layout(
    title="Top 10 Most Frequent Artists (2015–2024)",
    xaxis_title="Artist",
    yaxis_title="Mentions",
    hovermode="x unified",
    plot_bgcolor="black",
    paper_bgcolor="black",
    font=dict(color="white", size=14),
    hoverlabel=dict(bgcolor="#491d8b", font_size=14, font_color="white"),
    margin=dict(l=100, r=150, t=60, b=50),
    autosize=True
)

fig.show()
```

```python
# pitchfork_top_artists_bar.py

import pandas as pd
import re
from ast import literal_eval
import plotly.graph_objects as go

# Load dataset
df = pd.read_csv("/content/merged_2015_2024_with_wiki_images.csv")

# Filter for rows that have a Pitchfork rank and valid artist data
pitchfork_df = df.dropna(subset=["pitchfork_rank", "artists"])

# Map artists to songs specifically for Pitchfork
artist_song_map_pitchfork = {}

for _, row in pitchfork_df.iterrows():
    raw = row["artists"]
    song = row["song"]
    try:
        parsed = literal_eval(raw) if isinstance(raw, str) and raw.startswith("[") else [raw]
    except:
        parsed = [raw]

    for entry in parsed:
        if isinstance(entry, str):
            names = re.split(r"\s*,\s*|\s*/\s*|\s+and\s+", entry)
            for name in names:
                artist = name.strip()
                if artist:
                    if artist not in artist_song_map_pitchfork:
                        artist_song_map_pitchfork[artist] = {"count": 0, "songs": set()}
                    artist_song_map_pitchfork[artist]["count"] += 1
                    artist_song_map_pitchfork[artist]["songs"].add(song)
        elif isinstance(entry, list):
            for nested_artist in entry:
                if isinstance(nested_artist, str):
                    names = re.split(r"\s*,\s*|\s*/\s*|\s+and\s+", nested_artist)
                    for name in names:
                        artist = name.strip()
                        if artist:
                            if artist not in artist_song_map_pitchfork:
                                artist_song_map_pitchfork[artist] = {"count": 0, "songs": set()}
                            artist_song_map_pitchfork[artist]["count"] += 1
                            artist_song_map_pitchfork[artist]["songs"].add(song)

# Top 10 artists based on Pitchfork mentions
sorted_artists_pitchfork = sorted(artist_song_map_pitchfork.items(), key=lambda x: x[1]["count"], reverse=True)[:10]

artists_pitchfork = [a[0] for a in sorted_artists_pitchfork]
counts_pitchfork = [a[1]["count"] for a in sorted_artists_pitchfork]
hover_texts_pitchfork = [
    f"{a[0]}<br>Pitchfork Mentions: {a[1]['count']}<br>Songs:<br>{'<br>'.join(sorted(a[1]['songs']))}"
    for a in sorted_artists_pitchfork
]

fig_pitchfork = go.Figure(go.Bar(
    x=artists_pitchfork,
    y=counts_pitchfork,
    hovertext=hover_texts_pitchfork,
    hoverinfo="text",
    marker=dict(color="#1192e8")
))

fig_pitchfork.update_layout(
    title="Top 10 Most Frequent Artists on Pitchfork Lists (2015–2024)",
    xaxis_title="Artist",
    yaxis_title="Pitchfork Mentions",
    hovermode="x unified",
    plot_bgcolor="black",
    paper_bgcolor="black",
    font=dict(color="white", size=14),
    hoverlabel=dict(bgcolor="#003a6d", font_size=14, font_color="white"),
    margin=dict(l=100, r=150, t=60, b=50),
    autosize=True
)

fig_pitchfork.show()
```

```python
# songs_in_both_table.py

import pandas as pd
import plotly.graph_objects as go
from IPython.display import display, HTML

# Load data
df = pd.read_csv("merged_2015_2024_with_wiki_images.csv")

# Filter songs with both Pitchfork and Billboard ranks
songs_in_both = df.dropna(subset=['pitchfork_rank', 'billboard_rank']).copy()
songs_in_both['pitchfork_rank'] = songs_in_both['pitchfork_rank'].astype(int)
songs_in_both['billboard_rank'] = songs_in_both['billboard_rank'].astype(int)
songs_in_both['average_rank'] = ((songs_in_both['pitchfork_rank'] + songs_in_both['billboard_rank']) / 2).round(1)

result_df = songs_in_both[['year', 'artists', 'song', 'pitchfork_rank', 'billboard_rank', 'average_rank', 'image']]

# Generate HTML with hoverable image preview
html = '''
<style>
  body { background-color: #000; color: #fff; font-family: 'Courier New', monospace; }
  table { width: 100%; border-collapse: collapse; background: rgba(255,255,255,0.1); }
  th, td { padding: 10px; border-bottom: 1px solid rgba(255,255,255,0.2); text-align: left; font-family: 'Courier New', monospace; }
  th { background: rgba(255,255,255,0.2); }
  tr:hover { background-color: rgba(255, 255, 255, 0.3); }
  .preview-img { position: absolute; left: calc(100% - 240px); top: 0; display: none; width: 200px; border-radius: 12px; box-shadow: 0 0 20px rgba(255,255,255,0.3); z-index: 100; }
</style>
<h2>Songs Ranked on Both Pitchfork and Billboard</h2>
<table id="rank-table">
  <thead>
    <tr>
      <th>Year</th>
      <th>Artist</th>
      <th>Song</th>
      <th>Pitchfork Rank</th>
      <th>Billboard Rank</th>
      <th>Average Rank</th>
    </tr>
  </thead>
  <tbody>
'''

for i, row in result_df.iterrows():
    image_url = row['image'].replace('"', '')
    html += f'''
    <tr onmousemove="const img=document.getElementById('hover-img'); img.src='{image_url}'; img.style.top=event.clientY+'px'; img.style.display='block';"
        onmouseleave="document.getElementById('hover-img').style.display='none';">
      <td>{row['year']}</td>
      <td>{row['artists']}</td>
      <td>{row['song']}</td>
      <td>{row['pitchfork_rank']}</td>
      <td>{row['billboard_rank']}</td>
      <td>{row['average_rank']}</td>
    </tr>
    '''

html += '''</tbody></table><img id="hover-img" class="preview-img" />'''

# Display the interactive table
display(HTML(html))
```
